```
import nltk
nltk.download('punkt')

from sklearn import tree
from sklearn.feature_extraction import DictVectorizer
from sklearn.pipeline import Pipeline

from google.colab import drive
drive.mount('drive/')

import csv
```

```
# baca file test XOR train
def kalimat_text(file_name, jenis_data):
  jenis_data = jenis_data.lower()
  if jenis_data == 'train':
    panjang = 50
  else:
    if jenis_data == 'test':
      panjang = 10
    else:
      return False
# baca file
  tempat = []
  with open(file_name, 'r') as tsv:
    tsv_reader = csv.reader(tsv, delimiter='\n', quoting=csv.QUOTE_NONE)
    for row in tsv_reader:
      tempat.append(','.join(row))
  # misahin perkalimat
  dot = []
  for i in range(len(tempat)):
    if tempat[i] == '.\tZ':
      dot.append(i+1)
  # nentuin start mulai untuk lanjut
  start = [0]
  for i in range(panjang-1):
    start.append(dot[i])
  # buat relace sting sama jadiin biar jadi text per klaimat
  outer = []
  for i in range(panjang):
    inner = []
    for j in range(start[i],dot[i]):
      inner.append(tempat[j].replace('\t', '/'))
    outer.append(inner)
  alhamdulillah = []
  # ini jadi 1 array
  for i in range(len(outer)):
    alhamdulillah.append(' '.join(outer[i]))
  return alhamdulillah
```

```
# train
train = kalimat_text('drive/My Drive/Colab Notebooks/NLP/DataTrain/train.01.tsv', 'train')
print(train)
```

```
# test
test = kalimat_text('drive/My Drive/Colab Notebooks/NLP/DataTest/test_sentences.tsv', 'test')
print(test)
```

```
# misahin antara word dan tag
def sentences_tag(text):
  sentences = [] # list untuk menampung kalimat dan kata-kata di dalamnya
  tags = [] # list untuk menampung tag
  # split in tuh per pemisahnya
  for i in range(0,len(text)):
    tokens = text[i].split()
    sent = []
    tag = []
    # masttin kalo misal gada pemisahnya brrt digabunign sama kata depanya
    for j in range(0,len(tokens)):
      if '/' not in tokens[j]:
          tokens[j+1] = tokens[j]+' '+tokens[j+1]
      pair = tokens[j].split('/')
      # ini biar gada duplikat aja sih kara bekas gabungan tadi
      if len(pair) > 1:
        sent.append(pair[0])
        tag.append(pair[1])
    sentences.append(sent)
    tags.append(tag)
  return sentences, tags
```

```
# train
train_s_t = sentences_tag(train)
print(train_s_t)
```

```
# test
test_s_t = sentences_tag(test)
print(test_s_t)
```

```
# fungsi ekstraksi fitur
def features(sentence, index):
    """ sentence: [w1, w2, ...], index: the index of the word """
    print("sentence index = ")
    print(sentence[index])
    prefix_1 = ''
    prefix_2 = ''
    suffix_1 = ''
    suffix_2 = ''
    if (len(sentence[index])>2):
      prefix_1 = sentence[index][0]
      prefix_2 = sentence[index][:2]
      suffix_1 = sentence[index][-1]
      suffix_2 = sentence[index][-2:]
    return {
        'word': sentence[index],
        'prefix-1': prefix_1,
        'prefix-2': prefix_2,        
        'suffix-1': suffix_1,
        'suffix-2': suffix_2,        
        'prev_word': '' if index == 0 else sentence[index - 1],
        'next_word': '' if index == len(sentence) - 1 else sentence[index + 1],
    }
```

```
# transormasi dataset
def transform_to_dataset(sentences, tags):
    X, y = [], []
    for sentence_idx in range(len(sentences)):
        for index in range(len(sentences[sentence_idx])):
            X.append(features(sentences[sentence_idx], index))
            y.append(tags[sentence_idx][index])
    return X, y
```

```
# proses trainnig data train
X, y = transform_to_dataset(train_s_t[0], train_s_t[1])
from sklearn import tree
from sklearn.feature_extraction import DictVectorizer
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', DictVectorizer(sparse=False)),
    ('classifier', tree.DecisionTreeClassifier(criterion='entropy'))
])
clf.fit(X, y)   
print('Training completed')
```

```
# testing
X_test, y_test = transform_to_dataset(test_s_t[0], test_s_t[1])
```

```
print("Accuracy: ",clf.score(X_test, y_test)*100,'%')
```

```
from nltk import word_tokenize

def pos_tag(sentence):
    tags = clf.predict([features(sentence, index) for index in range(len(sentence))])
    #return zip(sentence, tags)
    return tags
 

cek = ''
cok = []
for i in test_s_t[0]:
  for j in i:
    if j not in cok:
      cok.append(j)
      cek += j+' '
print(pos_tag(word_tokenize(cek)))
```
